```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# Read in the data
datapath = "../data/diabetes_binary_5050split_health_indicators_BRFSS2015.csv.zip"
data_in = pd.read_csv(datapath, compression="zip")
data_in.head()
```

```python
# one split for train / test
# Separate the data into training and testing
data_train, data_test = train_test_split(
    data_in,
    test_size=0.2)
```

```python
# save to file
train_filepath = "../data/train/diabetes_binary_train.csv.zip"
data_train.to_csv(train_filepath, index=False)
test_filepath = "../data/test/diabetes_binary_test.csv.zip"
data_test.to_csv(test_filepath, index=False)
```
